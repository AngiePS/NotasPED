```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))

counts1, bins1, patches1 = ax1.hist(np, bins=20, rwidth=0.9, color=(0, 0, 0.55))
ax1.set_title("Notas finales parciales")
ax1.set_ylabel("Número de estudantes")
ax1.set_xlabel("Nota")

for count, patch in zip(counts1, patches1):
    if count > 0:
        ax1.text(patch.get_x() + patch.get_width() / 2, count, str(int(count)), ha='center', va='bottom')

counts2, bins2, patches2 = ax2.hist(ne, bins=20, rwidth=0.9, color=(0.678, 0.847, 0.902))
ax2.set_title("Notas Exame")
ax2.set_ylabel("Número de estudantes")
ax2.set_xlabel("Nota")

for count, patch in zip(counts2, patches2):
    if count > 0:
        ax2.text(patch.get_x() + patch.get_width() / 2, count, str(int(count)), ha='center', va='bottom')

counts3, bins3, patches3 = ax3.hist(nf, bins=20, rwidth=0.9, color=(0, 0.7, 0))
ax3.set_title("Notas Finales")
ax3.set_ylabel("Número de estudantes")
ax3.set_xlabel("Nota")

for count, patch in zip(counts3, patches3):
    if count > 0:
        ax3.text(patch.get_x() + patch.get_width() / 2, count, str(int(count)), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## Aprovados - Reprovados
